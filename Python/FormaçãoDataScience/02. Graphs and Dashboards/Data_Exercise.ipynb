Learning with libraries: matplotlib, pandas, numpy, and seaborn.

```python
#Importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
dtdados = pd.read_csv('z_dados.csv', sep= ';')
dtdados.head()
```

----
# GDP Visualization by Municipality 
- (PIB Column)

```python
#Checking the GDP (PIB column) of each municipality
municipio_pib = dtdados.groupby(['MUNICIPIO'])['PIB'].sum()
municipio_pib
```

```python
#Checking in graphical format
sns.histplot(dtdados.PIB, kde=True, bins=10).set_title('GDP')
```

```python
#Checking the 05 municipalities with the lowest GDPs in the dataset
low_pib = dtdados.sort_values('PIB', ascending= True).head()
low_pib = low_pib.iloc[:,1:3]
    # {:} All rows of the dataset 
    # {1:3} To view columns 1 through 3 (Python ignores the last column)
low_pib
```

```python
#Checking in graphical format
low_pib.plot.barh(x='MUNICIPIO',y='PIB', color= 'red')
```

```python
#Checking the 5 municipalities with the highest GDPs in the dataset
high_pib = dtdados.sort_values('PIB', ascending=False).head()
high_pib = high_pib.iloc[:,1:3]
    # {:} All rows of the dataset 
    # {1:3} To view columns 1 through 3 (Python ignores the last column)
high_pib
```

```python
#Checking in graphical format
high_pib.plot.barh(x='MUNICIPIO',y='PIB', color= 'blue')
```

# Commitment Visualization by Municipality 
- ("VALOREMPENHO" Column)

```python
#Checking the Commitment Value (VALOREMPENHO column) of each municipality
municipio_empenho = dtdados.groupby(['MUNICIPIO'])['VALOREMPENHO'].sum()
municipio_empenho
```

```python
#Checking in graphical format
sns.histplot(dtdados.VALOREMPENHO, kde=True, bins=10, color='blue').set_title('Commitment Value')
```

```python
#Checking the 05 municipalities with the lowest Commitment Value in the dataset
low_commitment = dtdados.sort_values('VALOREMPENHO', ascending=True).head()
low_commitment = low_commitment.iloc[:,[1,3]]
    # {:} All rows of the dataset 
    # {[1,3]} To visualize only the columns indexed at 1 and 3 (in this case 'MUNICIPIO' and 'VALOREMPENHO')
low_commitment
```

```python
#Checking in graphical format
low_commitment.plot.barh(x='MUNICIPIO', y='VALOREMPENHO', color='red')
```

```python
#Checking the 05 municipalities with the highest Commitment Value in the dataset
high_commitment = dtdados.sort_values('VALOREMPENHO', ascending= False).head()
high_commitment = high_commitment.iloc[:,[1,3]]
    # {:} All rows of the dataset 
    # {[1,3]} To visualize only the columns indexed at 1 and 3 (in this case 'MUNICIPIO' and 'VALOREMPENHO')
high_commitment
```

```python
#Checking in graphical format
high_commitment.plot.barh(x='MUNICIPIO', y='VALOREMPENHO', color='blue')
```
